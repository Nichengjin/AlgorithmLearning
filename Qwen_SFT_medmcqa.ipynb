```python
from trl import SFTConfig, SFTTrainer
from datasets import load_dataset
```

```python
dataset = load_dataset("openlifescienceai/medmcqa", split="train")
```

```python
training_args = SFTConfig(output_dir="Qwen/Qwen2.5-0.5B-SFT")
```

```python
def formatting_prompts_func(example):
    output_texts = []
    for i in range(len(example['question'])):
        text = f"### Question: {example['question'][i]} | Option A: {example['opa'][i]} | Option B: {example['opb'][i]} | Option C: {example['opc'][i]} | Option D: {example['opd'][i]}\n### Answer: {example['cop'][i]}"
        output_texts.append(text)
    return output_texts
```

```python
trainer = SFTTrainer(
    args=training_args,
    model="Qwen/Qwen2.5-0.5B",
    train_dataset=dataset,
    formatting_func=formatting_prompts_func
)
```

```python
trainer.train()
```
